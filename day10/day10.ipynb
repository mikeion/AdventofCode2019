```python
def set_of_asteroids(asteroids):
    set_of_asteroids=[]
    for i in range(len(asteroids)):
        for j in range(len(asteroids[i])):
            if asteroids[i][j] == "#":
                set_of_asteroids.append([i,j])
            else:
                continue
    return set_of_asteroids
```

```python
import numpy as np
import math
def best_asteroid(set_of_asteroids):
    dictionary_of_asteroids = {}
    for (x,y) in set_of_asteroids:
        count = 0
        potential_asteroids = [(i-x, j-y) for (i,j) in set_of_asteroids]
        generators = []
        for (c,d) in potential_asteroids:
            if math.gcd(c,d) == 1:
                generators.extend([(c,d),(c*2, d*2), (c*3, d*3), (c*4, d*4)])
                count +=1
            else:
                if (c,d) not in generators:
                    count +=1    
        dictionary_of_asteroids[(x,y)] = count
    return dictionary_of_asteroids
    #return dictionary_of_asteroids[max(dictionary_of_asteroids, key=dictionary_of_asteroids.get)]             
```

```python
best_asteroid(set_of_asteroids(test_asteroids))
```
