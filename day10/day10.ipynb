**Find the best location for a new monitoring station. How many other asteroids can be detected from that location?**

```python
asteroids = list(open('input.txt', 'r').read().split())
set_of_asteroids=[]
for i in range(len(asteroids)):
    for j in range(len(asteroids[i])):
        if asteroids[i][j] == "#":
            set_of_asteroids.append((i,j))
```

```python
# Drawing inspiration from my buddy Mauricio for this insight: https://github.com/mauricioinaz/

from math import atan2, pi

def angle_between_two_points(a,b):
    return atan2(b[0] - a[0], a[1] - b[1]) % (2 * pi)
```

```python
def seeable_asteroids(asteroids, asteroid_a):
    return len(set(angle_between_two_points(asteroid_a, asteroid_b) for asteroid_b in asteroids if asteroid_a != asteroid_b))
```

```python
def best_location(set_of_asteroids):
    return max(visible(set_of_asteroids, asteroid) for asteroid in set_of_asteroids)
```

```python
# Part One answer

print(best_location(set_of_asteroids))
```
